```python
import matplotlib.pyplot as plt
import pathlib
import numpy as np
from utils import read_im, save_im
output_dir = pathlib.Path("image_solutions")
output_dir.mkdir(exist_ok=True)
```

```python
im = read_im(pathlib.Path("images", "lake.jpg"))
plt.imshow(im)
```

## Task 2a


```python
def greyscale(im):
    """ Converts an RGB image to greyscale

    Args:
        im ([type]): [np.array of shape [H, W, 3]]

    Returns:
        im ([type]): [np.array of shape [H, W]]
    """

    R = 0.212
    G = 0.7152
    B = 0.0722

    for x in range(len(im)):
        for y in range(len(im[x])):
            im[x][y] = R * im[x][y][0] + G * im[x][y][1] + B * im[x][y][2]

    return im
```

```python
im_greyscale = greyscale(im)
save_im(output_dir.joinpath("lake_greyscale.jpg"), im_greyscale, cmap="gray")
plt.imshow(im_greyscale, cmap="gray")
```

## Task 2b


```python
def inverse(im):
    """ Finds the inverse of the greyscale image

    Args:
        im ([type]): [np.array of shape [H, W]]

    Returns:
        im ([type]): [np.array of shape [H, W]]
    """

    max_val = 1
    if im.dtype == np.uint8:
        max_val = 255

    for x in range(len(im)):
        for y in range(len(im[x])):
            im[x][y] = max_val - im[x][y]

    return im
```

```python
im_inverse = inverse(im)
save_im(output_dir.joinpath("lake_inverse.jpg"), im_inverse)
plt.imshow(im_inverse)
```
